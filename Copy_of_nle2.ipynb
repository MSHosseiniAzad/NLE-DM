```
#delete the name train from train.txt!!!
import os

# مسیر دایرکتوری حاوی تصاویر
images_dir = '/content/drive/MyDrive/bdd100k_images_10k_txt/bdd100k/images/10k_images'

# مسیر ذخیره فایل txt خروجی
out_txt = '/content/train_txt/train.txt'

# باز کردن فایل txt برای نوشتن
with open(out_txt, 'w') as f:

    # حلقه بر روی فایل‌های تصویر
    for filename in os.listdir(images_dir):
        # استخراج نام بدون پسوند
        name = os.path.splitext(filename)[0]

        # نوشتن نام در فایل txt
        f.write(name + '\n')
```
